```python
import os
import urllib.request
import zipfile
import tarfile

import matplotlib.pyplot as plt
%matplotlib inline
from PIL import Image
import numpy as np
import sklearn
```

```python
data_dir = "./data/"
if not os.path.exists(data_dir):
    os.mkdir(data_dir)
```

```python
from sklearn.datasets import fetch_openml

mnist = fetch_openml('mnist_784', version=1, data_home="./data/")
fasion = fetch_openml('Fashion-MNIST',data_home='./data/')
```

```python
X = mnist.data.to_numpy()
y = mnist.target.to_numpy()
```

```python
train_data_path = "./data/train_images"
if not os.path.exists(train_data_path):
    os.mkdir(train_data_path)
```

```python
# Get 7,8 images from data
image_0 = 0
image_7 = 0
image_8 = 0

for i in range(len(X)):
    if(y[i]=='0') and image_0<1000:
        image_path = f"./data/train_images/image_0_{image_0}.jpg"
        image = (X[i].reshape(28,28))
        pil_image = Image.fromarray(image.astype(np.uint8))
        pil_image = pil_image.resize((64,64), Image.BICUBIC)
        pil_image.save(image_path)
        image_0 += 1

    if(y[i]=='7') and image_7<1000:
        image_path = f"./data/train_images/image_7_{image_7}.jpg"
        image = (X[i].reshape(28,28))
        pil_image = Image.fromarray(image.astype(np.uint8))
        pil_image = pil_image.resize((64,64), Image.BICUBIC)
        pil_image.save(image_path)
        image_7 += 1

    if(y[i]=='8') and image_8<1000:
        image_path = f"./data/train_images/image_8_{image_8}.jpg"
        image = (X[i].reshape(28,28))
        pil_image = Image.fromarray(image.astype(np.uint8))
        pil_image = pil_image.resize((64,64), Image.BICUBIC)
        pil_image.save(image_path)
        image_8 += 1
```
